## B. Calculations on the sample: proportion, mean, variance

Use the cleaned numeric columns only where needed (drop rows with missing body mass or bill length, or use `.dropna(subset=[...])` as you prefer). Show your code and round printed numbers sensibly.

**1.** For the variable `body_mass_g`:

- Compute the **proportion** of penguins with body mass **greater than** the overall median body mass (among non-missing values).
- Compute the **mean** and **variance** of `body_mass_g` (same subset).

```python
body_mass = df["body_mass_g"].dropna()
median_body_mass = body_mass.median()
proportion_above_median = (body_mass > median_body_mass).mean()
mean_body_mass = body_mass.mean()
variance_body_mass = body_mass.var()

print(f"Median body mass: {median_body_mass:.0f} g")
print(f"Proportion above median: {proportion_above_median:.3f}")
print(f"Mean body mass: {mean_body_mass:.2f} g")
print(f"Variance of body mass: {variance_body_mass:.2f} g^2")
```

**2.** For `body_mass_g`, identify at least **two** distributional summary measures that describe **center** and **spread** (name them and compute them). Optionally add a third that describes **shape** (e.g. skewness), if you know how to compute it.

```python
body_mass = df["body_mass_g"].dropna()

mean_body_mass = body_mass.mean()
median_body_mass = body_mass.median()
std_body_mass = body_mass.std()
iqr_body_mass = body_mass.quantile(0.75) - body_mass.quantile(0.25)
skew_body_mass = body_mass.skew()

print(f"Center - mean: {mean_body_mass:.2f} g")
print(f"Center - median: {median_body_mass:.0f} g")
print(f"Spread - standard deviation: {std_body_mass:.2f} g")
print(f"Spread - IQR: {iqr_body_mass:.2f} g")
print(f"Shape - skewness: {skew_body_mass:.3f}")
```

```python
flipper = df["flipper_length_mm"].dropna()
mu = flipper.mean()
sigma = flipper.std()

print(f"Mean (mu): {mu:.2f} mm")
print(f"Standard deviation (sigma): {sigma:.2f} mm")
print()

for k in [1, 2, 3]:
    lower = mu - k * sigma
    upper = mu + k * sigma
    print(f"mu +/- {k} sigma: ({lower:.2f}, {upper:.2f}) mm")
```

## F. Outliers

**1.** Name **two** methods for identifying or **removing outliers** on a numeric column. For each, give a **one-sentence** explanation of the rule. Then show **one** method applied in code to flag outliers in `bill_length_mm` (do not delete them from the full dataset the count or indices is enough).

```python
bill_length = df["bill_length_mm"].dropna()

# Method 1: IQR rule
q1 = bill_length.quantile(0.25)
q3 = bill_length.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outlier_mask = (bill_length < lower_bound) | (bill_length > upper_bound)
outlier_indices = bill_length[outlier_mask].index.tolist()

print("Method 1: IQR rule flags values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR as outliers.")
print("Method 2: Z-score rule flags values whose standardized distance from the mean is larger than a chosen cutoff, often 3.")
print()
print(f"Lower bound: {lower_bound:.2f}")
print(f"Upper bound: {upper_bound:.2f}")
print(f"Number of flagged outliers in bill_length_mm: {len(outlier_indices)}")
print(f"Outlier indices: {outlier_indices}")
```
